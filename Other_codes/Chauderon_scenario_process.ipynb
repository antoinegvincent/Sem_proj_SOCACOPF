```python
import cvxpy as cp
import numpy as np
import math
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import gurobipy
```

```python
#Import data
scenario_df = pd.read_csv("Data_632_/Chaudron_scenarios.csv",delimiter=";")
load_df = pd.read_csv("Data_632_/Chauderon_load.csv",delimiter=";")
build_df = pd.read_csv("Data_632_/building_data_KNN.csv")

# Add the grid node to which eachbuilding is connected:
load_df = load_df.merge(build_df[["Hub","grid_node"]],on="Hub",how="right")

# Transform strings into float for processing
load_df['Domestic_electricity'] = load_df['Domestic_electricity'].str.replace(',', '.').astype(float)
load_df['SolarGains'] = load_df['SolarGains'].str.replace(',', '.').astype(float)
```

```python
# Get the scenarios to iterate
scenarios = scenario_df.Period.to_list()
```

```python
# create the final dataframe with the active power demand for each bus for each time step for each scenario
final_df = pd.DataFrame()
for sc in scenarios:
    intermediate_df = load_df[load_df.Period == sc].drop(columns=["Hub","Period"]).groupby(["Time","grid_node"]).sum().reset_index()
    intermediate_df["Period"] = sc
    cols = [intermediate_df.columns[-1]] + list(intermediate_df.columns[:-1])
    intermediate_df = intermediate_df[cols]
    final_df = pd.concat([final_df, intermediate_df], axis=0, ignore_index=True)
```

```python
final_df.to_csv("Data_632_/Clean_demand_632.csv")
```
